ESCENARIO  1

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)
# Inicializacin de la figura 3D
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
# Configuracin de colores y posiciones
colors = ['r', 'g', 'b', 'y']
yticks = [3, 2, 1, 0]
# Bucle para crear capas de barras en diferentes planos Y
for c, k in zip(colors, yticks):
  xs = np.arange(20)
  ys = np.random.rand(20)

# Definicin de colores por barra
  cs = [c] * len(xs)
  cs[0] = 'c'

#  Dibujo de las barras en 3D
  ax.bar(xs, ys, zs=k, zdir='y', color=cs, alpha=0.8)
# Etiquetas
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
# Configuracin de ticks del eje Y
ax.set_yticks(yticks)
#Mostrar grfico.
plt.show()
```

ESCENARIO 2

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from matplotlib.ticker import LinearLocator
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
# Crear datos.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Trazar la superficie.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,linewidth=0, antialiased=False)
# Personaliza el eje z.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Agregue una barra de color que asigne valores a colores.
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
