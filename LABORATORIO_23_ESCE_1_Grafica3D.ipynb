```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)
# Inicialización de la figura 3D
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
# Configuración de colores y posiciones
colors = ['r', 'g', 'b', 'y']
yticks = [3, 2, 1, 0]
# Bucle para crear capas de barras en diferentes planos Y
for c, k in zip(colors, yticks):
  xs = np.arange(20)
  ys = np.random.rand(20)

# Definición de colores por barra
  cs = [c] * len(xs)
  cs[0] = 'c'

#  Dibujo de las barras en 3D
  ax.bar(xs, ys, zs=k, zdir='y', color=cs, alpha=0.8)
# Etiquetas
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
# Configuración de ticks del eje Y
ax.set_yticks(yticks)
#Mostrar gráfico.
plt.show()
```
